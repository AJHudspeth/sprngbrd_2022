#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# Transpose dataframe and view indices

print(properties.T.index)
```

```python
# Transpose the dataframe and reset index 
properties = properties.T.reset_index()

# Assign the first row values to the columns and drop the redundant first row
first_row = properties.iloc[0]
print("\n The values of the first row: \n",first_row)

properties.columns=first_row
properties = properties.drop(0)
properties.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Rename first two columns, which were ostensibly 2-level column headers in the original excel file 
properties = properties.rename(columns={'Unnamed: 0':'Borough',pd.NaT:'Code'})
properties.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
props_melt = pd.melt(properties, id_vars=['Borough','Code']).rename(columns={0:'Month','value':'Average_price'})
props_melt.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Set types for object columns
props_melt=props_melt.astype({'Borough':'string','Code':'string','Average_price':'float'})

print(props_melt.dtypes)
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# Check the melted data frame to see if the ID (Code), Avg price, Borough and month have the same num els
props_melt.info()
```

```python
# Print the number of unique boroughs in the melted df
print('There are ' + str(props_melt['Borough'].nunique()) + ' unique values in the Borough column.')

# Display the unique elements of Borough 
print(props_melt['Borough'].value_counts())
```

```python
# Get a series of the London Boroughs, Comment off lines as needed to see pregression of steps

boroughs = (props_melt
            [~props_melt.Borough.str.contains('Unnamed')] # drop the unnamed boroughs with boolean mask
            ['Borough'] # pick off the Borough
            .value_counts() # perform value counts
            .rename_axis('region') # give the axis a dummy name
            .reset_index(name='counts') # make a nice little df with region and counts (won't need the counts col)
            ['region'][0:33] # pick off the first 33 elements (City of london and the 32 boroughs, see wiki to validate)
           ) # do not comment this paren

boroughs
```

```python
# Use the boroughs series from above to select only properties in the London Boroughs
props_bors = props_melt[props_melt.Borough.isin(boroughs)].dropna()
# Check to see that we have equal number of non-null
props_bors.info()
props_bors.head()
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# Choose Bor. Brent, for my Bro Brent
brent = props_bors[props_bors['Borough'] == 'Brent']

# Maybe we wanna look at sutton else, too
sutton = props_bors[props_bors['Borough'] == 'Sutton']

brent.head()
```

```python
# # Plot average price against month (not month against price)

fig, ax = plt.subplots()
# plot
brent.plot(x='Month',y='Average_price', kind='line', ax=ax)
sutton.plot(x='Month',y='Average_price', kind='line', ax=ax)
#props_bors.plot(x='Month',y='Average_price', kind='line')
plt.legend(['Brent', 'Sutton'])
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.title('Housing Prices of London Boroughs')
plt.show()
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:

```python
# groupby as implied in part 
dfg = (props_bors
       .groupby(['Borough','Year'])
       .mean() # take the average of prices by year and borough 
       .unstack() # swivel the inner index, year, to the columnspace
       # .reset_index() # idk that this is necessary, leave out to easily label bar plot with boroughs
      )
dfg.head()
```

```python
# Bar plots of Annual averages in '08 and '09 
# Housing crisis was '08, I thought, but the dip happened in '09
fig, ax = plt.subplots(figsize = (15,5))

(dfg
 [('Average_price', 2008)] # tuple indexing
 .plot(kind='bar', ax=ax, alpha=0.5, color="red")
)
(dfg
 [('Average_price', 2009)]
 .plot(kind='bar', ax=ax, alpha=1)
)
ax.set_ylabel('2008-2009 Housing Prices')
ax.legend(loc='best', bbox_to_anchor=(1,1))
ax.set_title('Housing Price Trends of London Boroughs')
```

```python
# Bar plots of Annual averages in 1995 and 2022 (newest and oldest data)
fig, ax = plt.subplots(figsize = (15,5))

(dfg
 [('Average_price', 2022)]
 .plot(kind='bar', ax=ax, alpha=0.5, color="red")
)
(dfg
 [('Average_price', 1995)]
 .plot(kind='bar', ax=ax, alpha=1)
)
ax.set_ylabel('Mean Housing Prices')
ax.legend(loc='best', bbox_to_anchor=(1,1))
ax.set_title('Housing Price Trends of London Boroughs')
```

```python
# dfg.columns
# Make some new columns

# the ratio of this year vs 20 years ago, and the difference of 
dfg[('Difference', '2022_over_2002')] = dfg[('Average_price', 2022)]/dfg[('Average_price', 2002)]

# the absolute difference between this year and 20 years ago
dfg[('Difference', '2022_minus_2020')] = dfg[('Average_price', 2022)]-dfg[('Average_price', 2002)]

# the ratio of 2018 prices to 1998, as instructed
dfg[('Difference', '2018_over_1998')] = dfg[('Average_price', 2018)]/dfg[('Average_price', 1998)]
```
